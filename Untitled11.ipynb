```python
import os
import json
import shutil

# Dataset ana yolu
dataset_root = "/content/drive/MyDrive/SARscope"

# ===============================
# TRAIN SET TAM KOPYALAMA
# ===============================
train_json_path = os.path.join(dataset_root, "train", "_annotations.coco.json")
train_image_dir = os.path.join(dataset_root, "train")

# Kopyalama yapılacak hedef klasör
full_train_root = os.path.join(dataset_root, "train_full")
full_train_image_dir = os.path.join(full_train_root)
full_train_json_path = os.path.join(full_train_root, "_annotations.coco.json")

# Klasör oluştur
os.makedirs(full_train_image_dir, exist_ok=True)

# JSON'u doğrudan kopyala
shutil.copy(train_json_path, full_train_json_path)

# Tüm görselleri kopyala
for file_name in os.listdir(train_image_dir):
    src = os.path.join(train_image_dir, file_name)
    dst = os.path.join(full_train_image_dir, file_name)
    if os.path.isfile(src):
        shutil.copy(src, dst)

# ===============================
# VALIDATION SET TAM KOPYALAMA
# ===============================
val_json_path = os.path.join(dataset_root, "valid", "_annotations.coco.json")
val_image_dir = os.path.join(dataset_root, "valid")

# Kopyalama yapılacak hedef klasör
full_val_root = os.path.join(dataset_root, "valid_full")
full_val_image_dir = os.path.join(full_val_root)
full_val_json_path = os.path.join(full_val_root, "_annotations.coco.json")

# Klasör oluştur
os.makedirs(full_val_image_dir, exist_ok=True)

# JSON'u doğrudan kopyala
shutil.copy(val_json_path, full_val_json_path)

# Tüm görselleri kopyala
for file_name in os.listdir(val_image_dir):
    src = os.path.join(val_image_dir, file_name)
    dst = os.path.join(full_val_image_dir, file_name)
    if os.path.isfile(src):
        shutil.copy(src, dst)

print("✅ Tüm veri seti (train + valid) tam olarak kopyalandı!")
```

```python
# ===============================
# YAML DOSYASI OLUŞTURMA
# ===============================
yaml_content = f"""
train: {full_train_root}
val: {full_val_root}

# COCO formatını kullanacağımız için yalnızca image klasörlerini gösteriyoruz
# _annotations.coco.json dosyaları bu klasörlerde bulunur

nc: 1
names: ["ship"]
"""

yaml_path = os.path.join(dataset_root, "ship_data.yaml")
with open(yaml_path, "w") as f:
    f.write(yaml_content)

print(f"✅ Tüm veri seti kopyalandı ve YAML dosyası oluşturuldu: {yaml_path}")
```

```python
from ultralytics import YOLO

# Modeli yükle
model = YOLO("yolov8s.pt")  # Önceden eğitilmiş YOLOv8 small modeli

# Eğitim
model.train(
    data="/content/drive/MyDrive/SARscope/ship_data.yaml",  # YAML dosyasının tam yolu
    epochs=3,               # Epoch sayısı
    imgsz=224,              # Görüntü boyutu
    batch=16,               # Batch size
    project="sar_ship_detection",  # Proje klasörü
    name="yolov8_ship",     # Kayıt alt klasör ismi
    save=True,              # Eğitilen modeli kaydet
    save_period=1,          # Her epoch sonunda kaydet
    workers=2,              # Data loader worker sayısı
    verbose=True,           # Ayrıntılı çıktı
    plots=True              # Eğitim grafikleri oluştur
)

print("✅ Eğitim tamamlandı!")
```

```python
!pip install ultralytics pycocotools --upgrade

import os
import matplotlib.pyplot as plt
from ultralytics import YOLO
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

# ==============================
# 1️⃣ Modeli yükle
# ==============================
model = YOLO("/content/sar_ship_detection/yolov8_ship/weights/best.pt")

# ==============================
# 2️⃣ Val setinde değerlendirme
# ==============================
results = model.val(
    data="/content/drive/MyDrive/SARscope/ship_data.yaml",
    split="val",            # ✅ test yerine val
    save_json=True,
    plots=True
)

print("📊 YOLOv8 Dahili Değerlendirme Sonuçları:")
print(results)

# ==============================
# 3️⃣ COCO API ile doğruluk hesaplama
# ==============================
gt_json = "/content/drive/MyDrive/SARscope/valid/_annotations.coco.json"
pred_json = "runs/detect/val/predictions.json"  # val çalışınca buraya kaydedilir

coco_gt = COCO(gt_json)
coco_dt = coco_gt.loadRes(pred_json)

coco_eval = COCOeval(coco_gt, coco_dt, iouType='bbox')
coco_eval.evaluate()
coco_eval.accumulate()
coco_eval.summarize()

# ==============================
# 4️⃣ Val görsellerinde tahmin görselleştirme
# ==============================
val_images_dir = "/content/drive/MyDrive/SARscope/valid/images"
image_files = [os.path.join(val_images_dir, f)
               for f in os.listdir(val_images_dir) if f.lower().endswith(('.jpg', '.png'))][:8]

plt.figure(figsize=(12, 12))
for i, img_path in enumerate(image_files, 1):
    res = model(img_path)
    res_plotted = res[0].plot()
    plt.subplot(4, 4, i)
    plt.imshow(res_plotted)
    plt.axis("off")
    plt.title(os.path.basename(img_path))

plt.tight_layout()
plt.show()
```

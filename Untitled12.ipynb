```python
# =========================================
# 📌 1. Gerekli Kütüphaneler
# =========================================
import os
import shutil
import json
import wandb
from ultralytics import YOLO
from wandb.integration.ultralytics import add_wandb_callback
```

```python
# =========================================
# 📌 2. Veri Setini Kopyalama (Tüm Veri)
# =========================================
dataset_root = "/content/drive/MyDrive/SARscope"

# ---- TRAIN kopyalama ----
train_src = os.path.join(dataset_root, "train")
train_dest = os.path.join(dataset_root, "train_full")
os.makedirs(train_dest, exist_ok=True)

shutil.copy(os.path.join(train_src, "_annotations.coco.json"),
            os.path.join(train_dest, "_annotations.coco.json"))

for file_name in os.listdir(train_src):
    src = os.path.join(train_src, file_name)
    dst = os.path.join(train_dest, file_name)
    if os.path.isfile(src) and file_name != "_annotations.coco.json":
        shutil.copy(src, dst)

# ---- VALID kopyalama ----
val_src = os.path.join(dataset_root, "valid")
val_dest = os.path.join(dataset_root, "valid_full")
os.makedirs(val_dest, exist_ok=True)

shutil.copy(os.path.join(val_src, "_annotations.coco.json"),
            os.path.join(val_dest, "_annotations.coco.json"))

for file_name in os.listdir(val_src):
    src = os.path.join(val_src, file_name)
    dst = os.path.join(val_dest, file_name)
    if os.path.isfile(src) and file_name != "_annotations.coco.json":
        shutil.copy(src, dst)

print("✅ Tüm veri seti kopyalandı!")
```

```python
# =========================================
# 📌 3. ship_data.yaml Oluşturma
# =========================================
yaml_content = f"""
train: {train_dest}
val: {val_dest}

# COCO formatı için yalnızca klasörleri gösteriyoruz
nc: 1
names: ["ship"]
"""

yaml_path = os.path.join(dataset_root, "ship_data.yaml")
with open(yaml_path, "w") as f:
    f.write(yaml_content)

print(f"✅ YAML dosyası oluşturuldu: {yaml_path}")
```

```python
# =========================================
# 📌 5. YOLOv11 Modelini Yükleme
# =========================================
model = YOLO("yolo11n.pt")  # İhtiyaca göre yolo11s.pt, yolo11m.pt kullanılabilir

# W&B callback ekle
add_wandb_callback(model, enable_model_checkpointing=True)
```

```python
# =========================================
# 📌 6. Model Eğitimi
# =========================================
model.train(
    data=yaml_path,       # YAML dosyasının tam yolu
    epochs=5,             # Epoch sayısı
    imgsz=640,            # Görüntü boyutu
    batch=16,             # Batch size
    project="sar_ship_detection_yolo11",
    name="yolo11_ship_run",
    save=True,
    save_period=1,
    workers=2,
    verbose=True,
    plots=True
)

print("✅ Eğitim tamamlandı!")
```

```python
# ===============================
# 1. Kurulum ve Kütüphaneler
# ===============================
!pip install ultralytics matplotlib scikit-learn --quiet

import os
import json
import torch
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score
from tqdm import tqdm
from ultralytics import YOLO

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"✅ Device: {device}")
```

```python
# ===============================
# 2. Parametreler
# ===============================
dataset_root = "/content/drive/MyDrive/ship_dataset_2"  # Senin dataset yolun
checkpoint_dir = "/content/drive/MyDrive/yolo12_checkpointsyolov12"
os.makedirs(checkpoint_dir, exist_ok=True)

num_epochs = 5          # Eğitim epoch sayısı
batch_size = 4          # GPU'ya göre ayarla
subset_ratio = 0.1      # %10 subset ile hızlı eğitim
iou_threshold = 0.5     # IoU eşik değeri

yaml_path = os.path.join(dataset_root, "data.yaml")
```

```python
import os
import random
import shutil

# ===============================
# Subset Oranı
# ===============================
subset_ratio = 0.1  # %10 subset
original_train_images = os.path.join(dataset_root, "train/images")
original_train_labels = os.path.join(dataset_root, "train/labels")  # YOLO formatında gerekirse
subset_train_dir = os.path.join(dataset_root, f"subset_train_{int(subset_ratio*100)}")

if not os.path.exists(subset_train_dir):
    os.makedirs(os.path.join(subset_train_dir, "images"), exist_ok=True)
    os.makedirs(os.path.join(subset_train_dir, "labels"), exist_ok=True)

    # Orijinal görüntü listesini al
    all_images = os.listdir(original_train_images)
    random.shuffle(all_images)

    subset_count = int(len(all_images) * subset_ratio)
    subset_images = all_images[:subset_count]

    for img_name in subset_images:
        # Image kopyala
        shutil.copy(os.path.join(original_train_images, img_name),
                    os.path.join(subset_train_dir, "images", img_name))

        # Label kopyala (aynı isimle .txt olmalı)
        label_name = img_name.replace(".jpg", ".txt").replace(".png", ".txt")
        if os.path.exists(os.path.join(original_train_labels, label_name)):
            shutil.copy(os.path.join(original_train_labels, label_name),
                        os.path.join(subset_train_dir, "labels", label_name))

print(f"✅ Subset dataset oluşturuldu: {subset_train_dir}")
```

```python
yaml_path = os.path.join(dataset_root, f"data_subset_{int(subset_ratio*100)}.yaml")

with open(yaml_path, "w") as f:
    f.write(f"""
train: {os.path.join(subset_train_dir, "images")}
val: {os.path.join(dataset_root, "valid/images")}
test: {os.path.join(dataset_root, "test/images")}

nc: 1
names: ['ship']
""")

print(f"✅ Subset YAML created: {yaml_path}")
```

```python
# ===============================
# 4. Model Yükleme (YOLOv12)
# ===============================
model = YOLO("yolo12n.pt").to(device)  # Nano sürüm (hızlı test için)

# ===============================
# 5. Eğitim (Resume Destekli)
# ===============================
results = model.train(
    data=yaml_path,
    epochs=num_epochs,
    imgsz=640,
    batch=batch_size,
    project=checkpoint_dir,
    name="yolo12_ship",
    resume=False
)
```

```python
# ===============================
# 6. Değerlendirme Fonksiyonları
# ===============================
def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])
    inter = max(0, x2 - x1) * max(0, y2 - y1)
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    union = area1 + area2 - inter
    return inter / union if union > 0 else 0

def evaluate_yolo(model, images_dir, ann_path, iou_threshold=0.5):
    preds = model.predict(source=images_dir, imgsz=640, conf=0.25, device=device)
    with open(ann_path, 'r') as f:
        gt_data = json.load(f)

    gt_images = {img['id']: img for img in gt_data['images']}
    gt_by_image = {}
    for ann in gt_data['annotations']:
        gt_by_image.setdefault(ann['image_id'], []).append(ann['bbox'])

    all_scores, all_labels, ious = [], [], []
    for pred in preds:
        img_id = None
        for k, v in gt_images.items():
            if v['file_name'] == os.path.basename(pred.path):
                img_id = k
                break
        if img_id is None:
            continue

        gt_boxes = gt_by_image.get(img_id, [])
        pred_boxes = pred.boxes.xyxy.tolist()
        pred_scores = pred.boxes.conf.tolist()

        for pb, score in zip(pred_boxes, pred_scores):
            best_iou = 0
            for gt in gt_boxes:
                gx1, gy1, gw, gh = gt
                gb = [gx1, gy1, gx1 + gw, gy1 + gh]
                best_iou = max(best_iou, compute_iou(pb, gb))

            all_scores.append(score)
            all_labels.append(1 if best_iou >= iou_threshold else 0)
            if best_iou >= iou_threshold:
                ious.append(best_iou)

    precision, recall, _ = precision_recall_curve(all_labels, all_scores)
    ap = average_precision_score(all_labels, all_scores)
    return precision, recall, ap, ious
```

```python
# ===============================
# 7. Eğitim Sonrası Değerlendirme
# ===============================
val_images = os.path.join(dataset_root, "valid/images")
val_ann = os.path.join(dataset_root, "valid/annotations.json")

precision, recall, ap, ious = evaluate_yolo(model, val_images, val_ann, iou_threshold)

print(f"\n✅ AP: {ap:.4f}")
print(f"✅ Ortalama IoU: {sum(ious)/len(ious) if ious else 0:.4f}")
```

```python
# ===============================
# 8. Grafikler
# ===============================
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(recall, precision, label=f'AP={ap:.4f}')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(ious, bins=20, range=(0, 1))
plt.xlabel("IoU")
plt.ylabel("Frequency")
plt.title("IoU Distribution")

plt.show()
```
